```python
import pandas as pd
from datasets import Dataset

# CSV 로드
df = pd.read_csv("0608_merged_kor_eng_sms.csv")  # 업로드한 파일명에 맞게 수정
df = df[["text", "label_encoded"]].dropna()
```

```python
# HuggingFace Dataset으로 변환
dataset = Dataset.from_pandas(df)
dataset = dataset.train_test_split(test_size=0.2, seed=42)
```

```python
df.shape
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from datasets import DatasetDict # Import DatasetDict
```

```python
model_name = "bert-base-multilingual-cased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
def tokenize(example):
    # Tokenize the text column
    tokenized_inputs = tokenizer(example["text"], truncation=True, padding="max_length", max_length=128)
    # The original 'label_encoded' column should still be present.
    # We don't need to add it here if the map function preserves it.
    return tokenized_inputs
```

```python
tokenized_dataset = dataset.map(tokenize, batched=True)

# Rename the 'label_encoded' column to 'labels' as expected by the Trainer
# Use DatasetDict's rename_column method on each split
tokenized_dataset = tokenized_dataset.rename_column("label_encoded", "labels")

# Set the format to PyTorch tensors
tokenized_dataset.set_format("torch", columns=["input_ids", "attention_mask", "labels"])

model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)
```

# ---

```python
import os
from transformers import TrainingArguments, Trainer
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix, balanced_accuracy_score
from transformers import AutoModelForSequenceClassification # Make sure this is imported

# Ensure AutoModelForSequenceClassification is imported if it wasn't in the original snippet
# from transformers import AutoModelForSequenceClassification

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    report = classification_report(labels, preds, output_dict=True, zero_division=0)
    bal_acc = balanced_accuracy_score(labels, preds)
    return {
        "accuracy": report["accuracy"],
        "f1_ham": report["0"]["f1-score"],
        "f1_spam": report["1"]["f1-score"],
        "balanced_accuracy": bal_acc
    }
args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    save_strategy="epoch",
    per_device_train_batch_size=16,
    per_device_eval_batch_size=32,
    num_train_epochs=3,
    logging_dir="./logs",
    load_best_model_at_end=True,
    # Explicitly disable reporting to prevent wandb errors
    report_to="none"
)

# os.environ["WANDB_DISABLED"] = "true" # This line can be kept or removed, but adding report_to is more reliable

# Assuming model, tokenized_dataset, and tokenizer are defined from previous cells
trainer = Trainer(
    model=model,
    args=args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["test"],
    compute_metrics=compute_metrics,
    tokenizer=tokenizer
)

trainer.train()
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# 모델과 토크나이저 로드
model_path = "0608_test_model"
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForSequenceClassification.from_pretrained(model_path)
model.eval()  # 평가모드

# 전처리 함수 (학습 때와 동일하게 유지)
import re
def clean_sms(text):
    text = text.lower()
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"[^\w\s가-힣]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text
```

```python
def predict_spam(text):
    # 텍스트 정제
    clean_text = clean_sms(text)

    # 토큰화
    inputs = tokenizer(clean_text, return_tensors="pt", truncation=True, padding="max_length", max_length=128)

    # 예측
    with torch.no_grad():
        outputs = model(**inputs)
        probs = torch.nn.functional.softmax(outputs.logits, dim=-1)
        pred_label = torch.argmax(probs, dim=1).item()

    label_str = "spam" if pred_label == 1 else "ham"
    spam_prob = probs[0][1].item()
    return label_str, spam_prob
```

```python
# 한 문장씩 입력받아 스팸 여부 판별
while True:
    msg = input("\n✏️ 분류할 문장을 입력하세요 (종료하려면 'exit'): ")

    if msg.lower() == "exit":
        print("🔚 예측 종료")
        break

    label, prob = predict_spam(msg)
    print(f"📬 예측 결과: {'🟥 스팸' if label == 'spam' else '🟩 정상 메시지'} | 스팸 확률: {prob:.2%}")
```

```python
# ✅ 1. 라이브러리 로드
import os
import numpy as np
import pandas as pd
import re
from datasets import Dataset, DatasetDict
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, balanced_accuracy_score
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    TrainingArguments,
    Trainer,
)

# ✅ 2. 데이터 로딩 (전처리 안 된 버전 기준)
df = pd.read_csv("0608_merged_kor_eng_sms.csv")  # label_encoded, text 포함

# ✅ 3. train/test 분할
train_df, test_df = train_test_split(df, test_size=0.2, stratify=df["label_encoded"], random_state=42)
```

```python
from sklearn.utils import resample

# train_df를 클래스별 분리
train_spam = train_df[train_df["label_encoded"] == 1]
train_ham = train_df[train_df["label_encoded"] == 0]

# spam을 ham 수만큼 복제 (또는 반대로 ham을 줄여도 됨)
train_spam_upsampled = resample(train_spam,
                                replace=True,
                                n_samples=len(train_ham),
                                random_state=42)

# 병합 후 셔플
train_balanced = pd.concat([train_ham, train_spam_upsampled]).sample(frac=1, random_state=42)

# 전처리 다시 적용
train_balanced["clean_text"] = train_balanced["text"].apply(clean_sms)
```

```python
# ✅ 4. 전처리 함수
def clean_sms(text):
    if not isinstance(text, str):
        return ""
    text = text.lower()
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"[^\w\s가-힣]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

# ✅ 5. 전처리 적용
train_balanced["clean_text"] = train_balanced["text"].apply(clean_sms)
test_df["clean_text"] = test_df["text"].apply(clean_sms)
```

```python
# ✅ 6. Hugging Face Dataset 변환
dataset = DatasetDict({
    "train": Dataset.from_pandas(train_balanced[["clean_text", "label_encoded"]]),
    "test": Dataset.from_pandas(test_df[["clean_text", "label_encoded"]])
})
```

```python
# ✅ 7. Tokenizer 준비 및 적용
model_name = "bert-base-multilingual-cased"
tokenizer = AutoTokenizer.from_pretrained(model_name)

def tokenize(example):
    return tokenizer(example["clean_text"], truncation=True, padding="max_length", max_length=128)

tokenized_dataset = dataset.map(tokenize, batched=True)
tokenized_dataset = tokenized_dataset.rename_column("label_encoded", "labels")
tokenized_dataset.set_format("torch", columns=["input_ids", "attention_mask", "labels"])

# ✅ 8. 모델 정의
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)

# ✅ 9. 평가 함수
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    report = classification_report(labels, preds, output_dict=True, zero_division=0)
    bal_acc = balanced_accuracy_score(labels, preds)
    return {
        "accuracy": report["accuracy"],
        "f1_ham": report["0"]["f1-score"],
        "f1_spam": report["1"]["f1-score"],
        "balanced_accuracy": bal_acc
    }

# ✅ 10. 학습 설정
args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    save_strategy="epoch",
    per_device_train_batch_size=16,
    per_device_eval_batch_size=32,
    num_train_epochs=3,
    load_best_model_at_end=True,
    report_to="none"
)

# ✅ 11. Trainer 구성 및 학습
trainer = Trainer(
    model=model,
    args=args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["test"],
    compute_metrics=compute_metrics,
    tokenizer=tokenizer
)
```

```python
trainer.train()
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# 모델과 토크나이저 로드
model_path = "0608_oversampling_saved_model"
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForSequenceClassification.from_pretrained(model_path)
model.eval()  # 평가모드

# 전처리 함수 (학습 때와 동일하게 유지)
import re
def clean_sms(text):
    text = text.lower()
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"[^\w\s가-힣]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text
```

```python
def predict_spam(text):
    # 텍스트 정제
    clean_text = clean_sms(text)

    # 토큰화
    inputs = tokenizer(clean_text, return_tensors="pt", truncation=True, padding="max_length", max_length=128)

    # 예측
    with torch.no_grad():
        outputs = model(**inputs)
        probs = torch.nn.functional.softmax(outputs.logits, dim=-1)
        pred_label = torch.argmax(probs, dim=1).item()

    label_str = "spam" if pred_label == 1 else "ham"
    spam_prob = probs[0][1].item()
    return label_str, spam_prob
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.utils import resample
```

```python
# 1. CSV 파일 로드
df = pd.read_csv("0608_merged_kor_eng_sms_dedup.csv")
```

```python
# 2. 언어 감지 함수 (한글 포함 여부로 구분)
def detect_language(text):
    if pd.isnull(text):
        return "unknown"
    return "korean" if any('\uac00' <= ch <= '\ud7a3' for ch in text) else "english"

df["language"] = df["text"].apply(detect_language)

# 3. label 통일 (정수형 라벨 생성)
df["label_str"] = df["label"].map({0: "ham", 1: "spam", "ham": "ham", "spam": "spam"})
df["label_encoded"] = df["label_str"].map({"ham": 0, "spam": 1})

# 4. train/test 분할
train_df, test_df = train_test_split(
    df,
    test_size=0.3,
    stratify=df["label_encoded"],
    random_state=42
)

# 5. 오버샘플링: 영어 spam
eng_spam = train_df[(train_df["language"] == "english") & (train_df["label_str"] == "spam")]
eng_spam_oversampled = resample(eng_spam,
                                replace=True,
                                n_samples=train_df[(train_df["language"] == "english") & (train_df["label_str"] == "ham")].shape[0],
                                random_state=42)

# 6. 다운샘플링: 한국어 ham
kor_ham = train_df[(train_df["language"] == "korean") & (train_df["label_str"] == "ham")]
kor_spam = train_df[(train_df["language"] == "korean") & (train_df["label_str"] == "spam")]

kor_ham_downsampled = resample(kor_ham,
                               replace=False,
                               n_samples=kor_spam.shape[0],
                               random_state=42)
```

```python
# 7. 나머지 포함하여 최종 train 구성
others = train_df[
    ~train_df.index.isin(eng_spam.index) &
    ~train_df.index.isin(kor_ham.index)
]

train_final = pd.concat([eng_spam_oversampled, kor_ham_downsampled, others], ignore_index=True).sample(frac=1, random_state=42)

# 8. 결과 출력 확인
print("🔍 최종 학습용 샘플 수:", len(train_final))
print(train_final["label_str"].value_counts())

# # (선택) 저장
# train_final.to_csv("train_balanced_kor_eng.csv", index=False, encoding="utf-8-sig")
# test_df.to_csv("test_original_kor_eng.csv", index=False, encoding="utf-8-sig")
# print("✅ 저장 완료: train_balanced_kor_eng.csv / test_original_kor_eng.csv")
```

```python
import os
from transformers import TrainingArguments, Trainer
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix, balanced_accuracy_score
from transformers import AutoModelForSequenceClassification # Make sure this is imported

# Ensure AutoModelForSequenceClassification is imported if it wasn't in the original snippet
# from transformers import AutoModelForSequenceClassification

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    report = classification_report(labels, preds, output_dict=True, zero_division=0)
    bal_acc = balanced_accuracy_score(labels, preds)
    return {
        "accuracy": report["accuracy"],
        "f1_ham": report["0"]["f1-score"],
        "f1_spam": report["1"]["f1-score"],
        "balanced_accuracy": bal_acc
    }

args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    save_strategy="epoch",
    per_device_train_batch_size=16,
    per_device_eval_batch_size=32,
    num_train_epochs=3,
    logging_dir="./logs",
    load_best_model_at_end=True,
    # Explicitly disable reporting to prevent wandb errors
    report_to="none"
)

# os.environ["WANDB_DISABLED"] = "true" # This line can be kept or removed, but adding report_to is more reliable

# Assuming model, tokenized_dataset, and tokenizer are defined from previous cells
trainer = Trainer(
    model=model,
    args=args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["test"],
    compute_metrics=compute_metrics,
    tokenizer=tokenizer
)
```

```python
# 한글 포함 여부로 언어 감지
def detect_language(text):
    if pd.isnull(text):
        return "unknown"
    return "korean" if any('\uac00' <= ch <= '\ud7a3' for ch in text) else "english"

# 언어별 분기된 전처리 함수
def clean_sms_by_language(text, language):
    if not isinstance(text, str):
        return ""

    if language == "korean":
        # 한국어: 특수문자, URL, 숫자 제거 등
        text = re.sub(r"http\S+|www\S+", "", text)              # URL 제거
        text = re.sub(r"[^\w\s가-힣]", "", text)                # 한글/숫자/영어 제외 특수문자 제거
        text = re.sub(r"\d+", "", text)                         # 숫자 제거
        text = re.sub(r"\s+", " ", text).strip()
        return text.lower()

    elif language == "english":
        # 영어: 부호/기호는 의미 있을 수 있어 덜 aggressive
        text = re.sub(r"http\S+|www\S+", "", text)
        text = re.sub(r"[^a-zA-Z0-9\s.,!?$%]", "", text)         # 알파벳/숫자/일부 기호 유지
        text = re.sub(r"\s+", " ", text).strip()
        return text.lower()

    return text  # fallback
```

```python
def predict_spam(text):
    # 텍스트 정제
    clean_text = clean_sms(text)

    # 토큰화
    inputs = tokenizer(clean_text, return_tensors="pt", truncation=True, padding="max_length", max_length=128)

    # 예측
    with torch.no_grad():
        outputs = model(**inputs)
        probs = torch.nn.functional.softmax(outputs.logits, dim=-1)
        pred_label = torch.argmax(probs, dim=1).item()

    label_str = "spam" if pred_label == 1 else "ham"
    spam_prob = probs[0][1].item()
    return label_str, spam_prob
```

```python
# ✅ 최종 스팸 예측 함수
def predict_spam(text):
    # 1. 언어 감지
    language = detect_language(text)

    # 2. 언어별 정제
    clean_text = clean_sms_by_language(text, language)

    # 3. 토큰화
    inputs = tokenizer(clean_text, return_tensors="pt", truncation=True, padding="max_length", max_length=128)

    # 4. 예측
    with torch.no_grad():
        outputs = model(**inputs)
        probs = torch.nn.functional.softmax(outputs.logits, dim=-1)
        pred_label = torch.argmax(probs, dim=1).item()

    # 5. 결과 출력
    label_str = "spam" if pred_label == 1 else "ham"
    spam_prob = probs[0][1].item()
    return label_str, spam_prob
```
